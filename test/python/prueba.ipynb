```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar CSV
df = pd.read_csv("../captures/log.csv")

# Crear figura
plt.figure(figsize=(12,6))

# Graficar los ángulos
plt.plot(df["timestamp"], df["roll"], label="roll")
plt.plot(df["timestamp"], df["pitch"], label="pitch")
plt.plot(df["timestamp"], df["yaw"], label="yaw")

plt.title("Euler Angles vs Timestamp")
plt.xlabel("Timestamp")
plt.ylabel("Angle (rad)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
